```python
from mup_transfer.architectures.activations import SigmaGeLU
import torch
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = torch.linspace(-10, 10, steps=100)
```

```python
sigmas = [0.001, 0.01, 0.1, 1.0, 10.0]
colors  = sns.color_palette("inferno", len(sigmas))
for i, sigma in enumerate(sigmas):
    act = SigmaGeLU(sigma=sigma)
    y = act(x)
    plt.plot(x, y, label=f"sigma={sigma}", color=colors[i])
plt.plot(x, torch.nn.functional.relu(x), label="relu", color="black", linestyle="--")
# plt.yscale('symlog')
plt.legend()
```
